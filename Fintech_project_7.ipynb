### Data Selection

```python
df_corr = df.corr()
```

```python
rank = df_corr['fraud_ind_0']# Retrieving the correlation coefficients per feature in relation to the feature class
df_rank = pd.DataFrame(rank) 
df_rank = np.abs(df_rank).sort_values(by='fraud_ind_0',ascending=False) # Ranking the absolute values of the coefficients
                                                                  # in descending order
df_rank.dropna(inplace=True) # Removing Missing Data (not a number)
```

```python
# First we build our train dataset
df_train_all = df[0:1141340]
df_train_1 = df_train_all[df_train_all['fraud_ind_0'] == 1] # We seperate the data which are the frauds and the no frauds
df_train_0 = df_train_all[df_train_all['fraud_ind_0'] == 0]
print('In this dataset, we have ' + str(len(df_train_1)) +" frauds so we need to take a similar number of non-fraud")

df_sample=df_train_0.sample(300)
df_train = df_train_1.append(df_sample) # We gather the frauds with the no frauds. 
df_train = df_train.sample(frac=1) # Then we mix our dataset
```

```python
X_train = df_train.drop(['fraud_ind_0'],axis=1) # We drop the features Time (useless), and the Class (label)
y_train = df_train['fraud_ind_0'] # We create our label
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
```

```python
df_test_all = df[1141341:]
X_test_all = df_test_all.drop(['fraud_ind_0'],axis=1)
y_test_all = df_test_all['fraud_ind_0']
X_test_all = np.asarray(X_test_all)
y_test_all = np.asarray(y_test_all)
```

```python
X_train_rank = df_train[df_rank.index[1:11]] # We take the first ten ranked features
X_train_rank = np.asarray(X_train_rank)
```

```python
X_test_all_rank = df_test_all[df_rank.index[1:11]]
X_test_all_rank = np.asarray(X_test_all_rank)
y_test_all = np.asarray(y_test_all)
```

### Confusion Matrix

```python
class_names=np.array(['0','1']) # Binary label, Class = 1 (fraud) and Class = 0 (no fraud)
```

```python
# Function to plot the confusion Matrix
def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd' 
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

### Model A

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, y_train) # Then we train our model, with our balanced data train.
```

```python
prediction_SVM_all = classifier.predict(X_test_all)
```

```python
cm = confusion_matrix(y_test_all, prediction_SVM_all)
plot_confusion_matrix(cm,class_names)
print("--------Model A,kernel=linear,weight=0.5:0.5--------")
print("TRAINING")
TN=cm[0][0]
FN=cm[0][1]
FP=cm[1][0]
TP=cm[1][1]
pre=TP/(TP+FP)
recall=TP/(TP+FN)
print("True negative:",TN)
print("False negative:",FN)
print("False Positive:",FP)
print("True Postive:",TP)
print("Precision:",pre)
print("Recall:",recall)
print("F-measure:",2*pre*recall/(pre+recall))
```

```python
classifier.fit(X_train_rank, y_train) # Then we train our model, with our balanced data train.
prediction_SVM = classifier.predict(X_test_all_rank) #And finally, we predict our data test.
```

```python
cm = confusion_matrix(y_test_all, prediction_SVM)
plot_confusion_matrix(cm,class_names)
print("--------Model A,kernel=linear,weight=0.5:0.5--------")
print("TEST")
TN=cm[0][0]
FN=cm[0][1]
FP=cm[1][0]
TP=cm[1][1]
pre=TP/(TP+FP)
recall=TP/(TP+FN)
print("True negative:",TN)
print("False negative:",FN)
print("False Positive:",FP)
print("True Postive:",TP)
print("Precision:",pre)
print("Recall:",recall)
print("F-measure:",2*pre*recall/(pre+recall))
```
